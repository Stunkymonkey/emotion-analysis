```python
#from IPython.display import Image
#Image(filename='inherit.png') 

#Hauptziel: Kompression!
#linguistisches vs. nicht-linguistisches Stemming

#Verfügbare Algorithmen im NLTK: 
#PorterStemmer: regelbasiert, simple Ersetzungsregeln, optimiert für Englisch (Porter, 1980)
#LancasterStemmer: implementiert den Paice/Husk Algorithmus (Paice, 1990)
#SnowballStemmer: andere Sprachen, implementiert aber im wesentlichen auch den Porter-Algorithmus 
#Regexp-Stemmer: erlaubt selbstdefinierte Ersetzungsregeln

#im NLTK nicht vorimplementiert: Lovins-Stemmer (Lovins, 1968)
```

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

example_words = ["python","pythoner","pythoning","pythoned","pythonly", "pythonic", "pythonian"]
 
for w in example_words:
    print(stemmer.stem(w))
```

```python
#Verwendungsbeispiel PorterStemmer 

from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

example_words = ['Such','an','analysis','can','reveal','features','that','are','not', 'easily', 'visible','from','the',
                  'variations', 'in', 'the','individual','genes','and','can','lead', 'to','a','picture', 'of','expression','that',
                 'is','more','biologically', 'transparent','and','accessible','to','interpretation', 'bogus']
 
for w in example_words:
    print(stemmer.stem(w), w)
```

```python
#Snowball-Stemmer: Porter für andere Sprachen!

from nltk.stem import SnowballStemmer

german_stemmer = SnowballStemmer("german")

example_words = ['Die','Computerlinguistik','beschäftigt','sich','mit','der',
                 'maschinellen','Verarbeitung','natürlicher',
                 'Sprache','Auf','Informatik', 'und','Linguistik','aufbauend','haben','sich',
                 'neue','und','eigenständige','Methoden','für',
                 'die','maschinelle','Verarbeitung',
                 'geschriebener','Sprache','entwickelt'] #vgl. Carstensen et. al., 2004

for w in example_words:
    print(german_stemmer.stem(w),w)

```

```python
#auch gelegentlich anzutreffende Unterscheidung: algorithmische Stemmer (wie oben) und ressourcenbasierte Stemmer (Lemmatizer)

#from IPython.display import Image
#Image(filename='stem_lemma.png') 
```

```python
#NLTK: WordNet-basierter Lemmatizer
#deutlich langsamer, da er versucht, jedes Token zu einem WordNet-Eintrag zu mappen
#existiert kein WordNet Eintrag, wird das Originaltoken zurückgegeben

from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

example_words = ['Such','an','analysis','can','reveal','features','that',
                 'are','not', 'easily', 'visible','from','the',
                  'variations', 'in', 'the','individual',
                 'genes','and','can','lead', 'to','a','picture', 'of','expression','that',
                 'is','more','biologically', 'transparent',
                 'and','accessible','to','interpretation', 'bogus']

for w in example_words:
    print(lemmatizer.lemmatize(w),w)
    
#Anstatt rein morphologische Wurzeln findet der WordNet-Lemmatizer eine tatsächlich valide "Wurzel" 
#--> Stemmer sind formbasiert, Lemmatizer sind bedeutungsbasiert!
```
